### Plots

```python
x = np.arange(0, 3 * np.pi, 0.1)
```

```python
y = 1 + 2 * x
```

```python
subplot(1, 2, 1)
plot(x, y, 'r+')

a = np.arange(3*np.pi, 0, -0.1)
b = 1 + -2 * a
subplot(1, 2, 2)
plot(a, b, 'bo')
show()
```
